```python
import datetime
class BikeRental:#It has duties of BikeRental shop.
    def __init__(self,stock=0):#default stock is 0.If we are intializing some stock value it will take that
        self.stock=stock
    def get_stock(self):
        print("The available bikes are:",self.stock)
    def rent_hour(self,n):
        if n < self.stock:
            current_time=datetime.datetime.now()#calculate curent time while giving for rent and then return current time
            print("You are taking {}  bikes for rent at {} hours".format(n, current_time.hour))
            self.stock-=n
            return current_time
        else:
            print("Requested bikes are {} but the available no of bikes are {}".format(n,self.stock))
    def rent_daily(self,n):
        if n < self.stock:
            current_time=datetime.datetime.now()
            print("You are taking {}  bikes for rent at {}  hours".format(n, current_time.hour))
            self.stock-=n
            return current_time
        else:
            print("Requested bikes are {} but the available no of bikes are {}".format(n,self.stock))
    def rent_weekly(self,n):
        if n < self.stock:
            current_time=datetime.datetime.now()
            print("You are taking {}  bikes for rent at {} hours".format(n, current_time.hour))
            self.stock-=n
            return current_time
        else:
            print("Requested bikes are {} but the available no of bikes are {}".format(n,self.stock))
    def return_bike(self,info):
        bikes,time,basis=info
        bill=0
        if time and bikes and basis:#calculatin bill by considering info
            self.stock+=bikes
            duration_rented=datetime.datetime.now() - time
            if basis == 5:
                bill = (duration_rented.seconds / 3600) * 5 *bikes
            elif basis == 20:
                bill =(duration_rented.days / 3600) * 20 * bikes
            elif basis == 60:
               bill = (duration_rented.days / 3600) * 60 * bikes
            if (bikes >= 3)and(bikes <= 5):
                print("""We are giving,Group Rental, a promotion that can include from 3 to 5 Rentals (of any type) with a 
    discount of 30% of the total price""")
                bill = bill * 0.7
                print("Bike has returned")
                print("bill is: ${}".format(bill))
                return bill
            else:
                return bill
            
        

class BikeRiders:#It has duties of Bike riders.
    def __init__(self):#defaultly assign values of everything(like requested bikes,returned bikes) from riders as 0
        self.bikes=0
        self.time=0 
        self.basis=0
        self.bill=0
    def requestingNoofBikes(self):
        n=int(input("Eneter no.of bikes you are requesting for rent:"))
        self.bikes=n
        return self.bikes
    def return_bike(self):#reurns the info of riders like requested time,basis,count of bikes
        if self.bikes and self.time and self.basis:
            return self.bikes,self.time,self.basis
        else:
            return None

def main():
    brentals=BikeRental(1000)#object created to BikeRental class.
    briders=BikeRiders()#object created to BikeRiders class.
    while 1:
        print("""WHAT ARE YOU REQUESTING FOR
        1. See available bikes in the shop
        2. Rent bikes on hourly basis $5 per hour.
        3. Rent bikes on daily basis $20 per day.
        4. Rent bikes on weekly basis $60 per week.
        5. Return Bike
        6.Nothing""")
        
        option=int(input("Please enter selected option number:"))
        if option==1:
            print("You have requested to check the no of available bikes for rent")
            brentals.get_stock()
        elif option==2:
            print("You have requested for a bike to rent on hourly basis $5 per hour")
            nbikes=briders.requestingNoofBikes()
            briders.time=brentals.rent_hour(nbikes)
            briders.basis=5# for hourly basis assigning value with its cost per hour
        elif option==3:
            print("You have requested for a bike to rent on daily basis $20 per day")
            nbikes=briders.requestingNoofBikes()
            briders.time=brentals.rent_daily(nbikes)
            briders.basis=20# for hourly basis assigning value with its cost per day
        elif option==4:
            print("You have requested for a bike to rent on weekly basis $60 per week")
            nbikes=briders.requestingNoofBikes()
            briders.time=brentals.rent_weekly(nbikes)
            briders.basis=60# for hourly basis assigning value with its cost per week
        elif option==5:
            print("Return request is processed")
            info=briders.return_bike()
            bikereturn=brentals.return_bike(info)
            print("Your total bill is:${} only".format(round(bikereturn)))
        elif option==6:
            break;

if __name__== "__main__":#calling the main fucntion only when we are running this file.
    main()
```
